Phase 1: Sign-to-Text (180 hours)
We are also called the "Neural Gestural Recognition & Semantic Mapping"

Sign Video --> Feature Extraction --> Deep Model --> Text 

Camera Input --> Frame Capture --> Hand + Pose Detection (MediaPipe) --> Feature Extraction (Keypoints) --> Sequence Model (LSTM / Transformer) --> Text Output --> Display on Screen

![Phase1.png](Folder_Image\Phase1.png)

This is the core concept.
OpenCV captures image/video frames.
Each frame is passed to MediaPipe.
MediaPipe converts frames into packets (data + timestamp).
These packets flow through calculators (small processing units).
Multiple calculators form subgraphs (mini pipelines).
Multiple subgraphs form a graph (full pipeline).
The graph processes the frame and outputs landmarks.
Finally, OpenCV displays the result using cv2.imshow().
